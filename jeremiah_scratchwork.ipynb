```python
uncleaned_traffic_violations.groupby(['Arrest Type']).size()
```

### let's see the distribution of warnings vs citations based on the arrest type

```python
import matplotlib.pyplot as plt

pd.crosstab(uncleaned_traffic_violations["Arrest Type"], uncleaned_traffic_violations["Violation Type"], normalize="index")

outcomes = uncleaned_traffic_violations.groupby("Arrest Type")["Violation Type"].value_counts(normalize=True).unstack()
outcomes[["Warning","Citation"]].plot(kind="bar", stacked=True, figsize=(8,4))
plt.title("Proportion of Warning vs Citation by Arrest Type")
plt.ylabel("Proportion of Stops")
plt.show()
```

### See if Alcohol is a factor in issuing a violation for other charges

```python
uncleaned_traffic_violations.groupby('Alcohol').size()
```

```python
uncleaned_traffic_violations[uncleaned_traffic_violations['Charge'].astype(str).str.startswith('21-902')].groupby('Alcohol').size()
```

a good amount of non dui charges have yes for alcohol, potentially there's a factor?

```python
not_dui = uncleaned_traffic_violations[~uncleaned_traffic_violations['Charge'].astype(str).str.startswith('21-902')]
not_dui.head()
```

```python
# get distribution of warning vs citation by alcohol

pd.crosstab(not_dui["Alcohol"], not_dui["Violation Type"], normalize="index")

outcomes = not_dui.groupby("Alcohol")["Violation Type"].value_counts(normalize=True).unstack()
outcomes[["Warning","Citation"]].plot(kind="bar", stacked=True, figsize=(8,4))
plt.title("Proportion of Warning vs Citation by Presence of Alcohol in Non-DUI related charges")
plt.ylabel("Proportion of Stops")
plt.show()
```
